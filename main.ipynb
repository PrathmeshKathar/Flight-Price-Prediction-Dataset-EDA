```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#feature engineering on Date_of_Journey

df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['month'] = df['Date_of_Journey'].str.split('/').str[1]
df['year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
# changing object datatype to int

df['Date'] = df['Date'].astype(int)
df['month'] = df['Date'].astype(int)
df['year'] = df['year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df['Arrival_time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Arrival_Hours'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minutes'] = df['Arrival_Time'].str.split(':').str[1]

df['Arrival_Minutes'] = df['Arrival_Minutes'].str.split(' ').str[0]  # also should do this, second option is  lambda function
```

```python
df.head(3)
```

```python
df['Arrival_Hours'] = df['Arrival_Hours'].astype(int)
df['Arrival_Minutes'] = df['Arrival_Minutes'].astype(int)
```

```python
df = df.drop('Arrival_Time',axis=1)
```

```python
df = df.drop('Arrival_time',axis=1)
```

```python
df.info()
```

```python
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_minute'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df['Dep_hour'] = df['Dep_hour'].astype(int)
df['Dep_minute'] =df['Dep_minute'].astype(int)
```

```python
df['Duration'].apply(lambda x:x.split(' ')[0])
```

```python
# categorical

df['Total_Stops'].mode()
```

```python
df['Total_Stops'].unique()
```

```python
# encoding also done(ordinal)!!
# imputetion is also done!!
# mapping with the help numbers

df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '2 stops':1, '1 stop':2, '3 stops':3,'4 stops':4,np.nan:1})
```

```python
df.head()
```

```python
# droping Route beacause of no work

df.drop('Route',axis=1,inplace=True)
```

```python
df.head()
```

##### work on Duration column beacause hours and minutes are mention.  also columns is in object data type

```python
df['Duration_hours'] = df['Duration'].str.split(' ').str[0]
```

```python
df['Duration_hours'] = df['Duration_hours'].str.split('h').str[0]
```

```python
df['Duration_minutes'] = df['Duration'].str.split(' ').str[1]
```

```python
df['Duration_minutes'] = df['Duration_minutes'].str.split('m').str[0]
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
#df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '2 stops':1, '1 stop':2, '3 stops':3,'4 stops':4,np.nan:1})
```

```python
# there is too much null values so we have to replace to with 0's
df['Duration_minutes'].isnull().sum()
```

```python
#replace with zero's done!!!

df['Duration_minutes'] = df['Duration_minutes'].fillna(0)
```

```python
# Convert to numeric and replace errors with NaN
df['Duration_minutes'] = pd.to_numeric(df['Duration_minutes'], errors='coerce').fillna(0).astype(int)
df['Duration_hours'] = pd.to_numeric(df['Duration_hours'], errors='coerce').fillna(0).astype(int)
```
